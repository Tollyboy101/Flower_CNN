```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
import pickle

IMG_SIZE = 75

img = cv2.imread("sunflower.jpg",cv2.IMREAD_GRAYSCALE)
```

```python
new_img = cv2.resize(img, (IMG_SIZE,IMG_SIZE))
```

```python
plt.imshow(new_img, cmap='gray')
```

```python
new_img = np.array(new_img).reshape(-1,IMG_SIZE,IMG_SIZE,1)
```

```python
pickle1 = open("test.pickle","wb")
pickle.dump(new_img,pickle1)
pickle1.close()
```
